```python
#Open file
fs = open('/Users/cmdb/data/genomes/BDGP6.Ensembl.81.gtf', 'r')

counter = 0
#Store the data in a list
p_genes = []
for i, line in enumerate(fs):
    #1. skip first 5 lines (header)
    if i < 5:
        continue
    #2. split the line into a list
    fields = line.split()
    #3. chrm, start, end, gene_name, feature_type, gene_biotype
    chrm = fields[0]
    start = fields[3]
    end = fields[4]
    #4. ignore entries not on 3R
    if chrm == '3R':
        #5. ignore non "gene" entries
        if fields[2] == 'gene':
            feature_type = fields[2]
            #6. ignore non-protein-encoding genes
            if 'protein_coding' in fields[-1]:
                #7. store gene_name and gene_biotype
                gene_name = fields[13]
                gene_biotype = fields[-1]
                p_genes += [(chrm, start, end, feature_type, gene_name, gene_biotype)]
                counter += 1
                #if counter < 10:
                    #print(chrm, start, end, feature_type, gene_name, gene_biotype)
                    #print(type(start))
                    
                    
#print('len of p_genes is', len(p_genes))

#binary search through p_genes list:

search_pos = 21378950
#initialize low, high, and mid
low = 0
high = len(p_genes)-1
mid = (len(p_genes))//2

gene = ''

search_tally = 0
while True:
    #print()
    #print('low is index', low)
    #print('high is index', high)
    #print('mid is index', mid)
    #print('search_pos is', search_pos)
    #print('tally is', search_tally)
    
    #print('p_genes[mid][1] is', p_genes[mid][1])
    #print('p_genes[mid][2] is', p_genes[mid][2])
        
    if high == low + 1:
        #print('high == low + 1')
        #find distance to end of low
        distance_to_low = search_pos - int(p_genes[low][2])
        #print('distance_to_low is', distance_to_low)
        #find distance to start of high
        distance_to_high = int(p_genes[high][1]) - search_pos
        #print('distance_to_high is', distance_to_high)
        
        #compare, figure out which is closest (i.e., which has smaller distance)
        if distance_to_low < distance_to_high:
            #low is the gene!
            #print('low is the gene')
            #print('the gene is', p_genes[low][4])
            print(p_genes[low][0:3], p_genes[low][4])
            lin_dist = distance_to_low
            
        else:
            #high is the gene!
            #print('high is the gene')
            #print(p_genes[high][4])
            print(p_genes[high][0:3], p_genes[high][4])
            lin_dist = distance_to_low
        
        break
        
    if search_pos < int(p_genes[mid][1]):
        high = mid
        #print('search_pos is less than mid')
        #print('high is now mid:', mid)
        mid = (high+low)//2
        #print('mid is now', mid)
    elif search_pos > int(p_genes[mid][2]):
        low = mid
        #print('search_pos is greater than mid')
        #print('low is now mid:', low)
        mid = (high+low)//2
        #print('mid is now', mid)
        
    else:
        #we found the gene!
        #output gene_name
        print('the gene is', p_genes[mid][4])
        break
    search_tally += 1
    #print('~~~~~~~~~~~~~~~~')
    
print('Linear genomic distance is', lin_dist)
print('search tally is', search_tally)
        
#Report the gene's linear genomic distance from 3R:21,378,950.
#How many iterations (a tally), did it take the search procedure 
#to find the nearest gene? Report this.
        
```
