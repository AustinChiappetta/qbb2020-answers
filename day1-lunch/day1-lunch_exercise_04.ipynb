#Question 4: Plot K4, K9, and K27 width distributions on the same plot as subplots

#Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bed_columns = ["chr", "start", "end"]
bed_columns
```

#Load data

```python
df_k4 = pd.read_table( "../../qbb2020/data/K4me3.bed", 
                     names=bed_columns)
df_k4
```

```python
df_k9 = pd.read_table( "../../qbb2020/data/K9me3.bed", 
                     names=bed_columns)
df_k9
```

```python
df_k27 = pd.read_table( "../../qbb2020/data/K27me3.bed", 
                     names=bed_columns)
df_k27
```

#Make width column for each

```python
df_k4.loc[:,"width"] = df_k4.loc[:,"end"] - df_k4.loc[:,"start"]
df_k4
```

```python
df_k9.loc[:,"width"] = df_k9.loc[:,"end"] - df_k9.loc[:,"start"]
df_k9
```

```python
df_k27.loc[:,"width"] = df_k27.loc[:,"end"] - df_k27.loc[:,"start"]
df_k27
```

#Define chromosomes of interest

```python
coi = ["2L", "2R", "3L", "3R", "4", "X"]
coi
```

#Plot

```python
fig, ax = plt.subplots(nrows=3, figsize=(10,10))
fig.suptitle("Width Distributions")

ax[0].hist(df_k4.loc[:,"width"])
ax[0].set_xlabel("Width (nucleotides)")
ax[0].set_ylabel("Number of metyhlated regions")
ax[0].set_title("K4me3 width distribution")

ax[1].hist(df_k9.loc[:,"width"])
ax[1].set_xlabel("Width (nucleotides)")
ax[1].set_ylabel("Number of metyhlated regions")
ax[1].set_title("K9me3 width distribution")

ax[2].hist(df_k27.loc[:,"width"])
ax[2].set_xlabel("Width (nucleotides)")
ax[2].set_ylabel("Number of metyhlated regions")
ax[2].set_title("K27me3 width distribution")

fig.tight_layout(pad=5.0)
```
