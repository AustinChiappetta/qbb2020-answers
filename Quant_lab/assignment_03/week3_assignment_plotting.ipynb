# 4. For each phenotype, produce a QQ plot and Manhattan plot. For the Manhattan plot, highlight SNPs with p-values less than 10-5 in a different color

```python
'/Users/cmdb/qbb2020-answers/Quant_lab/assignment_03/data/GWAS/file_list.txt'
```

```python
import numpy as np
#Get condition labels
pheno_labels = open('/Users/cmdb/qbb2020-answers/Quant_lab/assignment_03/data/pheno_labels.txt', 'r')
for line in pheno_labels:
    labels = line.split()

for num, label in enumerate(labels, start=1):
    gwas = pd.read_csv("/Users/cmdb/qbb2020-answers/Quant_lab/assignment_03/data/GWAS/plink.P" + str(num) + ".assoc.linear", sep = "\s+")

    
# split columns on any number of spaces
#Make Manhattan plots
    gwas['logP'] = -1 * np.log10(gwas['P'])
    gwas['snp_index'] = range(len(gwas))



    gwas['snp_index'] = range(len(gwas))

    fig, ax = plt.subplots()

    for chr in range(1, 23): 
        ax.scatter(gwas["snp_index"][gwas["CHR"] == chr], gwas["logP"][gwas["CHR"] == chr], marker = '.')

    plt.xlabel("SNPs")
    plt.ylabel("-log10(p-value)")
    ax.set_title("GWAS for " + label + " phenotype")
    
    plt.show()
```
