# 1. Visualize genetic relatedness between the strains by performing principal component analysis and plotting the first two components

```python
import pandas as pd
import matplotlib.pyplot as plt

PCA_file = '/Users/cmdb/qbb2020-answers/Quant_lab/assignment_03/data/plink.eigenvec'

```

```python
pca_output = pd.read_csv(PCA_file,
                      sep = '\t')
pca_output
```

```python
pca_output.columns = ['Family_ID', 'Sample_ID', 'PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10', 'PC11', 'PC12', 'PC13', 'PC14', 'PC15', 'PC16', 'PC17', 'PC18', 'PC19', 'PC20']
pca_output
```

```python
fig, ax = plt.subplots()
ax.scatter(x = pca_output['PC1'], y = pca_output['PC2'])
ax.set_title("PCA to Visualize Genetic Relatedness Between Strains")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
plt.show()
```

# 2. Visualize the allele frequency spectrum by plotting a histogram of allele frequencies

```python
allele_freqs = '/Users/cmdb/qbb2020-answers/Quant_lab/assignment_03/data/plink.frq'
df_freqs = pd.read_csv(allele_freqs, sep = '\s+')
df_freqs
```

```python
fig, ax = plt.subplots()

ax.hist(x = df_freqs["MAF"], log = True)
ax.set_ylabel("Counts")
ax.set_xlabel("Allele Frequency")
ax.set_title("Allele Frequency Spectrum")
fig.show()
```
