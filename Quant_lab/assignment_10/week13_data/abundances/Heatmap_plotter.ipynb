```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
abundances_file = '/Users/cmdb/qbb2020-answers/Quant_lab/assignment_10/week13_data/abundances/abundance_table.tab'
```

```python
abundances_df = pd.read_csv(abundances_file, sep = '\t')
abundances_df
```

```python
coi = ['Genomic bins', 'SRR492183', 'SRR492186', 'SRR492188', 'SRR492189', 'SRR492190', 'SRR492193', 'SRR492194', 'SRR492197']
subdf = abundances_df[coi]
subdf
```

```python
subdf = subdf.set_index('Genomic bins')
subdf = subdf.sort_index()
subdf
```

```python
subdf['Taxonomy_Annotations'] = ['Staphylococcus haemolyticus', 'Leuconostoc citreum', 
                                 'Staphylococcus lugdunensis', 'Enterococcus faecalis', 
                                 'Cutibacterium avidum', 'Staphylococcus epidermidis', 
                                 'Staphylococcus aureus', 'Anaerococcus prevotii']
subdf = subdf.set_index('Taxonomy_Annotations')
subdf
```

```python
subdf_log = np.log2(subdf)
subdf_log
```

```python
ax = plt.axes()
sns.heatmap(subdf, ax = ax)

ax.set_title('Genome Abundances')
plt.show()
```

```python
ax = plt.axes()
sns.heatmap(subdf_log, ax = ax)

ax.set_title('Genome Abundances (log2-transformed)')
plt.show()
```
